```python
# First import the basic libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

load the data set

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston
```

Features are the independent variables that will be used to predict the price. Target are the answers in predicting house prices, the data is what we are looking for.

```python
# load everything into a data frame

df_x=pd.DataFrame(boston.data,columns=boston.feature_names)
df_y=pd.DataFrame(boston.target)
```

Describe the data frame

```python
df_x.describe()
```

The mean age is 68.574901

```python
reg=linear_model.LinearRegression()
```

Split our training data  into training and testing data to validate the training model

```python
x_train, x_test, y_train,y_test = train_test_split(df_x,df_y,test_size=0.2,random_state=4)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

The coefficients are optimal

```python
a=reg.predict(x_test)
```

```python
a
```

If we evaluate the predictions with the actual values you find there is an error which can be resolved if we manipulate the parameters and fit to the predictions four or more times.

```python
#mean square error

np.mean((a-y_test)**2)
```
